## The code above creates a class of 1000 unique people.
## The code below creates a dataframe out of the class of 1000 unique people.

```python
import pandas as pd
people_df = pd.DataFrame.from_records([vars(p) for p in people_list])
```

```python
people_df.head()
```

```python
def format_phone_number(number):
    return '({}{}{}) {}{}{}-{}{}{}{}'.format(*number)

people_df['phone_num'] = people_df['phone_num'].astype(str)

people_df['phone_num'] = people_df['phone_num'].apply(format_phone_number)
```

```python
people_df.head()
```

## Now I'm going to create a people order dataframe

```python
# Create empty order data list
order_data = []

# Number of total orders
total_orders = 7894

# Number of unique customers
unique_customers = len(people_df)

# Number of orders per customer
orders_per_customer = total_orders // unique_customers

# Add orders for each customer
for i in range(unique_customers):
    name, id_num = random.choices(people_df['name'])[0], random.choices(people_df['id_num'])[0]
    for j in range(orders_per_customer):
        order_data.append([name, id_num, random.randint(1, 5), random.uniform(10, 100)])

# Create order dataframe
order_df = pd.DataFrame(order_data, columns=['name', 'id_num', 'Order Quantity', 'Order Amount'])
```
